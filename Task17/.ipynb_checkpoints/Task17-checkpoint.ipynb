## Sub plots

```python
import matplotlib.pyplot as plt

x = range(10)
y1 = [2, 4, 6, 8, 10, 8, 6, 4, 2, 0]
y2 = [0, 1, 4, 9, 16, 9, 4, 1, 0, 1]
categories = ['A', 'B', 'C', 'D']
values = [5, 10, 15, 20]
data1 = [2, 4, 3, 5, 4, 6, 7, 5, 4, 5]
data2 = [1, 3, 2, 4, 3, 5, 6, 4, 3, 4]

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
axs[0, 0].plot(x, y1, label='Line 1', color='blue')
axs[0, 0].plot(x, y2, label='Line 2', color='red')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].legend()

# Scatter plot
axs[0, 1].scatter(x, y1, color='green', label='Data 1')
axs[0, 1].scatter(x, y2, color='purple', label='Data 2')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].legend()

# Bar plot
axs[1, 0].bar(categories, values, color='skyblue')
axs[1, 0].set_title('Bar Plot')
axs[1, 0].set_xlabel('Categories')
axs[1, 0].set_ylabel('Values')

# Histogram
axs[1, 1].hist(data1, bins=5, alpha=0.5, label='Data 1')
axs[1, 1].hist(data2, bins=5, alpha=0.5, label='Data 2')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

## Histograms

```python
data1 = np.random.randn(1000)
data2 = np.random.normal(loc=1.0, scale=0.5, size=1000)  

plt.hist(data1, bins=30, alpha=0.5, label='Data 1')
plt.hist(data2, bins=30, alpha=0.5, label='Data 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Multiple Histograms')
plt.legend()
plt.show()
```

```python
plt.hist(data1, bins=30, density=True, alpha=0.5, label='Data 1')
plt.hist(data2, bins=30, density=True, alpha=0.5, label='Data 2')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normalized Histograms')
plt.legend()
plt.show()
```

## Box Plots

```python
# Sample data
data = np.random.randn(100, 5)

plt.boxplot(data)
plt.xlabel('Features')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```

```python
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=2, scale=1.5, size=100)
data3 = np.random.normal(loc=-2, scale=0.5, size=100)

plt.boxplot([data1, data2, data3], labels=['Data 1', 'Data 2', 'Data 3'])
plt.xlabel('Data Sets')
plt.ylabel('Value')
plt.title('Multiple Box Plots')
plt.show()
```
