# Task 7: Library Management System

```python
class Book:
    def __init__(self, title, author, ISBN, status="available"):
        self.title = title
        self.author = author
        self.ISBN = ISBN
        self.status = status
    
    def disp_info(self):
        print("Book name:", self.title)
        print("Author:", self.author)
        print("ISBN:", self.ISBN)
        print("Status:", self.status)
    
    def mark_as_borrowed(self):
        self.status = 'borrowed'

    def mark_as_returned(self):
        self.status = 'available'

class Member:
    def __init__(self, name, member_ID):
        self.name = name
        self.member_ID = member_ID
        self.borrowed_books = []
    
    def disp_info(self):
        print("Name: ", self.name)
        print("Member ID: ", self.member_ID)

    def borrow_book(self, book):
        if book.status == "available":
            self.borrowed_books.append(book)
            book.mark_as_borrowed()
            print(f"{self.name} borrowed {book.title}")
        else:
            print("Book is not available")
    
    def return_book(self, book):
        if book in self.borrowed_books:
            book.mark_as_returned()
            self.borrowed_books.remove(book)
            print(f"{self.name} returned {book.title}")
        else:
            print(f"{self.name} does not have {book.title} borrowed")

class Library:
    def __init__(self):
        self.bookList = []
        self.memberList = []
    
    def add_book(self, book):
        self.bookList.append(book)
        print(f"Book '{book.title}' added")
    
    def register_member(self, member):
        self.memberList.append(member)
        print(f"Member '{member.name}' added")
    
    def issue_book(self, member_ID, ISBN):
        
        member = None
        for m in self.memberList:
            if m.member_ID == member_ID:
                member = m
                break
        
        book = None
        for b in self.bookList:
            if b.ISBN == ISBN:
                book = b
                break
        
        if member and book:
            member.borrow_book(book)
        else:
            print("Member or Book not found")
    
    def restore_book(self, member_ID, ISBN):
        member = None
        for m in self.memberList:
            if m.member_ID == member_ID:
                member = m
                break
        
        book = None
        for b in self.bookList:
            if b.ISBN == ISBN:
                book = b
                break
        
        if member and book:
            member.return_book(book)
        else:
            print("Member or Book not found")
    
    def disp_books(self):
        for book in self.bookList:
            book.disp_info()

    def disp_members(self):
        for member in self.memberList:
            member.disp_info()

library = Library()
book1 = Book("Harry Potter", "J. K. Rowling", "1234")
book2 = Book("StillHouse Lake", "Rachel Caine", "5678")
book3 = Book("The Hobbit", "J. R. R. Tolkein", "5234")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

member1 = Member("Alex", "1")
member2 = Member("Brian", "2")
member3 = Member("Charlie", "3")

print("\n")
library.register_member(member1)
library.register_member(member2)
library.register_member(member3)

print("\n")
library.disp_books()
print("\n")
library.disp_members()

print("\n")
library.issue_book("1", "1234")  
library.issue_book("2", "5678") 
library.issue_book("3", "5234") 
print("\n")
library.disp_books()

print("\n")
library.restore_book("1", "1234") 
library.restore_book("2", "5678") 
print("\n")
library.disp_books()

```
