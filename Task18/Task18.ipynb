## Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

dataset = sns.load_dataset("iris")
sns.scatterplot(data=dataset, x="sepal_length", y="sepal_width", hue="species")
plt.title('Scatter Plot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

## Customized

```python
sns.set_style("whitegrid")
sns.scatterplot(data=dataset, x="sepal_length", y="sepal_width", hue="species")
plt.title('Scatter Plot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

## Line Plot

```python
data = sns.load_dataset("flights")

sns.lineplot(data=data, x="year", y="passengers", hue="month")
plt.title('Line Plot of Passengers Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

## Customized

```python
sns.set_style("dark")
sns.lineplot(data=data, x="year", y="passengers", hue="month")
plt.title('Line Plot of Passengers Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

## Bar plot

```python
data = sns.load_dataset("tips")

sns.barplot(data=data, x="day", y="total_bill", hue="sex")
plt.title('Bar Plot of Total Bill by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.set_style("dark")
sns.barplot(data=data, x="day", y="total_bill", hue="sex", palette="coolwarm")
plt.title('Bar Plot of Total Bill by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

## Count Plot

```python
data = sns.load_dataset("tips")

sns.countplot(data=data, x="day", hue="sex")
plt.title('Count Plot of Tips by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

## Customized

```python
sns.set_style("white")
sns.countplot(data=data, x="day", hue="sex", palette="deep")
plt.title('Count Plot of Tips by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

## Box plot

```python
data = sns.load_dataset("diamonds")
sns.boxplot(data=data, x="cut", y="price")
plt.title('Box Plot of Diamond Prices by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```

## Customized

```python
sns.boxplot(data=data, x="cut", y="price", hue='cut', palette='Set2')
plt.title('Box Plot of Diamond Prices by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```
