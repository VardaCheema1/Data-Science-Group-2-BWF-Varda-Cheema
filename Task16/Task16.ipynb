## Line Graphs

```python
import matplotlib.pyplot as plt


x = [1, 3, 5, 7, 9]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Multiple Lines Plot')
plt.legend()
plt.show()
```

```python
x = [1, 3, 5, 7, 9]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, marker='o')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot with Markers')
plt.show()
```

```python
x = [1, 3, 5, 7, 9]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 4, 5, 9, 13]
plt.plot(x, y1, color='red', linestyle='--', linewidth=2, label='Dashed Line')
plt.plot(x, y2, color='blue', linestyle='-', linewidth=2, label='Solid Line')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Customized Line Styles and Colors')
plt.legend()
plt.show()
```

## Scatter Plots

```python
x = [1, 3, 5, 7, 9]
y = [2, 4, 6, 8, 10]
plt.scatter(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Basic Scatter Plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
sizes = [20, 50, 100, 200, 500]
colors = ['red', 'green', 'blue', 'orange', 'purple']

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Customized Scatter Plot')
plt.show()
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 6, 8, 10]

plt.scatter(x1, y1, color='red', label='Dataset 1')
plt.scatter(x2, y2, color='blue', label='Dataset 2')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Multiple Scatter Plots')
plt.legend()
plt.show()
```

## Bar Charts

```python
categories = ['Cars', 'Trucks', 'Cycles', 'Jeeps', 'Motorbikes']
values = [4, 7, 1, 8, 5]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Chart')
plt.show()
```

```python
categories = ['Cars', 'Trucks', 'Cycles', 'Jeeps', 'Motorbikes']
values = [4, 7, 1, 8, 5]

plt.barh(categories, values)
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.show()
```

```python
categories = ['Cars', 'Trucks', 'Cycles', 'Jeeps', 'Motorbikes']
values = [4, 7, 1, 8, 5]

plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Chart')
plt.show()
```

## Pie Charts

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels)
plt.title('Basic Pie Chart')
plt.show()
```

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'pink', 'skyblue']

plt.pie(sizes, labels=labels, colors=colors)
plt.title('Pie Chart with Custom Colors and Legend')
plt.legend(labels, loc='best')
plt.show()
```

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'pink', 'skyblue']
explode = (0, 0, 0, 0.1)  

plt.pie(sizes, labels=labels, colors=colors, explode=explode)
plt.title('Exploded Slice Pie Chart')
plt.legend(labels, loc='best')
plt.show()
```

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'pink', 'skyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart with Percentage Labels')
plt.legend(labels, loc='best')
plt.show()
```
