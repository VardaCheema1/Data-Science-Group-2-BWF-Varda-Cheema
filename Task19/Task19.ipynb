## Linear Regression Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")
sns.regplot(data=data, x="total_bill", y="tip")
plt.title('Linear Regression Plot')
plt.show()
```

## Multiple Regression Plot

```python
sns.lmplot(data=data, x="total_bill", y="tip", hue="sex")
plt.title('Multiple Linear Regression Plot')
plt.show()
```

## Residual Plot

```python
sns.residplot(data=data, x="total_bill", y="tip")
plt.title('Residual Plot')
plt.show()
```

## Distribution
## Histogram

```python
sns.histplot(data=data, x="total_bill", kde=True)
plt.title('Histogram')
plt.show()
```

## Kernel Density Estimate Plot

```python
sns.kdeplot(data=data, x="total_bill", fill=True)
plt.title('KDE Plot')
plt.show()
```

## Distribution Plot

```python
sns.displot(data["total_bill"], kde=True)
plt.title('Distribution Plot')
plt.show()
```

## Box Plot

```python
sns.boxplot(data=data, x="day", y="total_bill")
plt.title('Box Plot')
plt.show()
```

## Violin Plot

```python
sns.violinplot(data=data, x="day", y="total_bill")
plt.title('Violin Plot')
plt.show()
```

## Pair Plot

```python
sns.pairplot(data)
plt.title('Pair Plot')
plt.show()
```

## Heat map

```python
flights = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")
sns.heatmap(flights, annot=True, fmt="d")
plt.title('Heatmap')
plt.show()
```

## Joint Plot

```python
sns.jointplot(data=data, x="total_bill", y="tip", kind="reg")
plt.title('Joint Plot')
plt.show()
```

## Facet Grid

```python
g = sns.FacetGrid(data, col="sex", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip").add_legend()
plt.title('Facet Grid')
plt.show()
```
